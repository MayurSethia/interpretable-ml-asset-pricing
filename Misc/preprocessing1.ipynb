```python
# 1. Forward/Backward fill or median impute for remaining columns

# GrSaleToGrOverhead: forward fill within each firm (grouped by 'permno')
# This is a ratio, so it might be reasonable to assume that the value does not drastically change over time
# and can be carried forward or backward within the same firm.
df1980['GrSaleToGrOverhead'] = df1980.groupby('permno')['GrSaleToGrOverhead'].ffill().bfill()

# LRreversal: forward fill within each firm (grouped by 'permno')
# This column could represent firm-specific trends over time, so filling it based on historical values makes sense.
df1980['LRreversal'] = df1980.groupby('permno')['LRreversal'].ffill().bfill()

# DebtIssuance: set to 0
# If the missing values in 'DebtIssuance' are not crucial or if we assume no debt issuance when the data is missing, 
# setting them to 0 is reasonable. This prevents any misrepresentation of the data with artificial values.
df1980['DebtIssuance'] = df1980['DebtIssuance'].fillna(0)

# cfp: set to 0 (assuming itâ€™s a non-essential metric for missing data)
# 'cfp' could represent cash flow per share or similar, and missing values could indicate the absence of relevant data. 
# Setting it to 0 avoids skewing the analysis and can be reasonable depending on its usage context.
df1980['cfp'] = df1980['cfp'].fillna(0)

# roaq: fill missing values with median
# 'roaq' represents return on average equity, which is a financial metric. 
# Using median imputation helps ensure that missing values are replaced by a central tendency, 
# avoiding the distortion that might arise from filling with 0 or forward/backward filling.
df1980['roaq'] = df1980['roaq'].fillna(df1980['roaq'].median())

# AbnormalAccruals: forward fill within each firm (grouped by 'permno')
# This could represent firm-specific accruals that change over time, so forward/backward filling within each firm
# is appropriate to capture the trend and reduce missing data.
df1980['AbnormalAccruals'] = df1980.groupby('permno')['AbnormalAccruals'].ffill().bfill()

# PctTotAcc: forward fill or set to 0 (depending on context)
# If this column is a time-series indicator for a firm's total accrual percentage, forward/backward filling makes sense.
# If its absence is not crucial to the analysis, setting to 0 might also be acceptable.
df1980['PctTotAcc'] = df1980.groupby('permno')['PctTotAcc'].ffill().bfill()

# InvestPPEInv: forward fill within each firm (grouped by 'permno')
# 'InvestPPEInv' likely refers to investments in property, plant, and equipment, which are typically firm-specific.
# Forward/backward filling ensures that missing values are filled in a way that makes sense over time for each firm.
df1980['InvestPPEInv'] = df1980.groupby('permno')['InvestPPEInv'].ffill().bfill()

# GP: fill missing values with median
# 'GP' represents gross profit, a key financial metric. Median imputation helps to replace missing values without
# introducing bias or outliers, which might distort the analysis if filled with zeros or forward/backward values.
df1980['GP'] = df1980['GP'].fillna(df1980['GP'].median())

# NumEarnIncrease: fill missing values with median
# 'NumEarnIncrease' represents the number of earnings increases for a firm. 
# Using the median ensures that the missing values are filled by the central tendency, 
# preventing distortion and reflecting the typical value in the dataset.
df1980['NumEarnIncrease'] = df1980['NumEarnIncrease'].fillna(df1980['NumEarnIncrease'].median())

# 2. Check for remaining missing values after imputation
# After performing the necessary imputation steps, we check the remaining missing values to ensure the dataset is complete
missing_pct_cleaned = df1980.isnull().mean() * 100

# Print the remaining missing values to confirm the impact of our imputation strategy
print("Remaining missing value % (top 10):")
print(missing_pct_cleaned.sort_values(ascending=False).head(10))
```

```python
import pandas as pd

# -----------------------------
# 1. Drop Columns with Excessive Missing or Limited Relevance
# -----------------------------
drop_cols = [
    'BetaTailRisk', 'MomOffSeason06YrPlus', 'MomSeason06YrPlus',
    'CBOperProf', 'EntMult', 'EarningsSurprise', 'GrSaleToGrInv',
    'HerfBE', 'HerfAsset', 'VarCF', 'ChInvIA'
]
df1980.drop(columns=drop_cols, inplace=True)

# -----------------------------
# 2. Sort Data by Firm and Time (important for forward/backward fill)
# -----------------------------
df1980.sort_values(by=['permno', 'yyyymm'], inplace=True)

# -----------------------------
# 3. Fill with Zero (Assuming Absence of Activity)
# -----------------------------
cols_fill_zero = ['ShareIss5Y', 'EarnSupBig', 'grcapx', 'Investment', 'DebtIssuance', 'cfp']
df1980[cols_fill_zero] = df1980[cols_fill_zero].fillna(0)

# -----------------------------
# 4. Median Imputation (for Key Financial Metrics)
# -----------------------------
median_impute_cols = [
    'BookLeverage', 'roaq', 'GP', 'NumEarnIncrease', 'Herf',
    'CashProd', 'NetDebtFinance', 'BPEBM', 'Leverage', 'SP', 'CF', 'BM'
]
for col in median_impute_cols:
    df1980[col] = df1980[col].fillna(df1980[col].median())

# -----------------------------
# 5. Forward/Backward Fill (Firm-Level Time Series Trends)
# -----------------------------
ffill_cols = [
    'IndRetBig', 'ChAssetTurnover', 'Price', 'GrSaleToGrOverhead', 'LRreversal',
    'AbnormalAccruals', 'PctTotAcc', 'InvestPPEInv', 'EBM', 'MomOffSeason', 'MomSeason'
]
for col in ffill_cols:
    df1980[col] = df1980.groupby('permno')[col].ffill().bfill()

# -----------------------------
# 6. Final Missing Value Check
# -----------------------------
missing_summary = df1980.isnull().mean().sort_values(ascending=False) * 100
print("Remaining missing value % (top 10):")
print(missing_summary.head(10))
```
